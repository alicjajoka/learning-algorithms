```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('images/bokeh.jpg')
```

```python
dst = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 10, 10)

plt.figure(figsize=(18,9))
plt.subplot(211), plt.imshow(img)
plt.subplot(212), plt.imshow(dst)
plt.show()
```

```python
dst = cv2.fastNlMeansDenoisingColored(img, None, 20, 20, 20, 20)

plt.figure(figsize=(18,9))
plt.subplot(211), plt.imshow(img)
plt.subplot(212), plt.imshow(dst)
plt.show()
```
