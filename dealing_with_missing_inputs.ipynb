**Wypełnianie brakujących danych jedną zadaną wartością**

```python
df = pd.read_csv("datasets/employees.csv", na_values = missing_inputs)
df.fillna({'Salary': 0, 'Gender': 'No gender'}).head(10)
```

**Wypełnianie brakujących danych wartością z poprzedniego wiersza**

```python
df.fillna(method = 'ffill').head(10)
```

**Wypełnianie brakujących danych wartością z następnego wiersza**

```python
df.fillna(method = 'bfill').head(10)
```

**Wypełnianie brakujacych danych medianą / średnią danych kolumn**

```python
df.fillna({'Salary': df['Salary'].median()}).head(10)
```

```python
df.fillna({'Salary': df['Salary'].mean()}).head(10)
```

```python
import random

df = pd.read_csv("datasets/employees.csv", na_values = missing_inputs)

g = df['Gender'].unique()
g = g[~pd.isnull(g)]
t = df['Team'].unique()
t = t[~pd.isnull(t)]

for index, row in df.iterrows():
    if pd.isnull(row['Gender']):
        df.fillna({'Gender': g[random.randint(0,len(g) - 1)]}, limit = 1, inplace = True)
    if pd.isnull(row['Team']):
        df.fillna({'Team': t[random.randint(0,len(t) - 1)]}, limit = 1, inplace = True)

df.head(20)
```
