```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df = pd.read_csv('datasets/advertising.csv')
df.head(10)
```

```python
fig = plt.figure(figsize=(24, 12))

ax1 = fig.add_subplot(131)
plt.scatter(df['TV'], df['Sales'])
plt.title('TV', fontdict = {'fontsize':32})

ax2 = fig.add_subplot(132)
plt.scatter(df['Radio'], df['Sales'])
plt.title('Radio', fontdict = {'fontsize':32})

ax3 = fig.add_subplot(133)
plt.scatter(df['Newspaper'], df['Sales'])
plt.title('Newspaper', fontdict = {'fontsize':32})

plt.show()
```

```python
sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=7, aspect=0.7, kind='reg')
```

```python
X = df.drop('Sales', axis=1)
y = df['Sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred = linreg.predict(X_test)

list(zip(y_test, y_pred))
```

```python
(metrics.mean_squared_error(y_test, y_pred)) ** (0.5)
```
